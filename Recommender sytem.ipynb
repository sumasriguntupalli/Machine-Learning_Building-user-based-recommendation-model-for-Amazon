## Recommendation model 

```python
import surprise
from surprise import Reader
from surprise import accuracy
from surprise import Dataset
from surprise.model_selection import train_test_split
from surprise import SVD
from surprise.model_selection import cross_validate
```

```python
movies_rating_melt = movies_rating.melt(id_vars = movies_rating.columns[0],value_vars=movies_rating.columns[1:],var_name="Movies",value_name="Rating")
```

```python
movies_rating_melt
```

```python
rd = Reader(rating_scale=(-1,10))
data = Dataset.load_from_df(movies_rating_melt.fillna(0),reader=rd)
data
```

```python
trainset, testset = train_test_split(data,test_size=0.25)
```

```python
svd=SVD()
svd.fit(trainset)
```

```python
pred = svd.test(testset)
```

```python
accuracy.rmse(pred)
```

```python
accuracy.mae(pred)
```

```python
cross_validate(svd, data, measures = ['RMSE', 'MAE'], cv = 3, verbose = True)
```

```python
def repeat(recom_type,dframe):
    rd = Reader()
    data = Dataset.load_from_df(dframe,reader=rd)
    print(cross_validate(recom_type, data, measures = ['RMSE', 'MAE'], cv = 3, verbose = True))
    print("--"*10)
    usr_id = 'A3R5OBKS7OM2IR'
    mv = 'Movie1'
    r_u = 5.0
    print(recom_type.predict(usr_id,mv,r_ui = r_u,verbose=True))
    print("--"*10)
```

```python
repeat(SVD(),movies_rating_melt.fillna(movies_rating_melt['Rating'].mean()))
```

```python
repeat(SVD(),movies_rating_melt.fillna(movies_rating_melt['Rating'].median()))
```

```python
#trying grid search and find optimum hyperparameter value for n_factors
from surprise.model_selection import GridSearchCV
```

```python
param_grid = {'n_epochs':[20,30],
             'lr_all':[0.005,0.001],
             'n_factors':[50,100]}
```

```python
gs = GridSearchCV(SVD,param_grid,measures=['rmse','mae'],cv=3)
data1 = Dataset.load_from_df(movies_rating_melt.fillna(movies_rating_melt['Rating'].median()),reader=rd)
gs.fit(data1)
```

```python
gs.best_score
```

```python
print(gs.best_score["rmse"])
print(gs.best_params["rmse"])
```
